```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x="Gender")
plt.title("Gender Distribution")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
gp=df.groupby('ParentEduc').agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
print(gp)
```

```python
sns.heatmap(gp,annot=True)
plt.title("Relationship between parent education and Student score")
plt.show()
```

# from the above analysis we cvan say that the education of parents has a major impact
# on the grades of their child for example the parents those has masters degree their child 
# perfoemance is hoghest among all the available data!!!
